```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for percentage of schools with water facility
water_data = {
    "Year": [2013, 2014, 2015, 2016],
    "Percentage": [78.2, 79.1, 80.3, 81.6]
}

# Data for gross enrollment ratio
enrollment_data = {
    "Year": [2013, 2014, 2015, 2016],
    "Enrollment Ratio": [0.87, 0.88, 0.89, 0.90]
}

# Convert data to pandas DataFrames
water_df = pd.DataFrame(water_data)
enrollment_df = pd.DataFrame(enrollment_data)

# Merge the two DataFrames
merged_df = water_df.merge(enrollment_df, on='Year')

# Calculate the correlation coefficient
correlation_coefficient = merged_df['Percentage'].corr(merged_df['Enrollment Ratio'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Percentage'], merged_df['Enrollment Ratio'], c='blue')
plt.xlabel('Percentage of Schools with Water Facility')
plt.ylabel('Gross Enrollment Ratio')
plt.title('Increase of GRE with Rise in % of schools with waterfacility')

# Add a trendline
from scipy import stats
m, b, r, p, se = stats.linregress(merged_df['Percentage'], merged_df['Enrollment Ratio'])
plt.plot(merged_df['Percentage'], m * merged_df['Percentage'] + b, color='red', label=f'Trendline (r={r:.2f})')

# Display the correlation coefficient
plt.annotate(f"Correlation Coefficient: {correlation_coefficient:.2f}", xy=(75, 0.92), fontsize=10, color='blue')

# Show the plot
plt.legend()
plt.grid(True)
plt.show()
```
